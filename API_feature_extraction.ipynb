```python
def fixer(line):
    try:
        arr = line.split("[")
        first = arr[0]
        middle = arr[1].split("]")[0].replace(",", " ")
        end = arr[1].split("]")[1]
        wline = first + middle + end + '\n'
    except IndexError:
        wline = line
    return wline
```

```python
# fix train 
linecount = 0
with open('api_features_train.dat') as infile:
    with open('api_features_train.2.csv','a') as outfile:
        for line in infile:
            outfile.write(fixer(line))
            linecount += 1
linecount
```

```python
def nn_fixer(line):
    arr = line.split(',')
    qid = arr[0]
    tags = ' '.join(arr[1:-2])
    label1, label2 = arr[-2:][0], arr[-2:][1]
    newline = qid + ',' + tags + ',' + label1 + ',' + label2
    return newline
```

```python
with open("nn_features_train.dat", "r") as infile:
    with open("nn_features_train.fixed.dat", "a") as outfile:
        for line in infile:
            outfile.write(nn_fixer(line))
```

```python
with open("nn_features_val.dat", "r") as infile:
    with open("nn_features_val.csv", "a") as outfile:
        for line in infile:
            outfile.write(nn_fixer(line))
```
